```python
%matplotlib inline 
directory = os.path.join('TCGA', 'validation')

for filename in os.scandir(directory):
    if filename.is_file():
      img = os.path.join(filename)
      results = model(img)
      plt.imshow(np.squeeze(results.render()))
      plt.savefig(f'results/{filename}.png')
      plt.close()
```
